```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
# IMPORTANT to set, so that the plots could be shown within the jupyter notebook
%matplotlib inline 

from matplotlib.pylab import rcParams
# Set the size for all the plots/figures
rcParams['figure.figsize'] = 14,5

# One can adjust the resolution for the plot, to better show the results on the screen
rcParams['figure.dpi'] = 300
rcParams['savefig.dpi'] = 300
```

```python
from sklearn import datasets

# import the dataset of iris for the following demo
iris = datasets.load_iris()

df_iris = pd.DataFrame(iris.data, columns=iris.feature_names)
df_iris['target'] = iris['target']

df_iris.head()
```

```python
# rename the columns
df_iris.rename(columns={'sepal length (cm)': 'sepal_length',
                        'sepal width (cm)': 'sepal_width',
                        'petal length (cm)': 'petal_length',
                        'petal width (cm)': 'petal_width'}, inplace=True)
```

## Line Plot

```python
# the minimal statement to plot all the columns within the dataframe as lines (by default)
ax = df_iris[['sepal_length', 'sepal_width']].plot()


ax.set_xlabel('samples')
ax.set_ylabel('attribute value (cm)')

# ignore the return value, to hide the print text on top of the plot
_ = ax.set_title('Sample Title')
```

## Bar Plot

```python
# the minimal statement to print all the columns within the dataframe (by default)
ax = df_iris[['target']].plot(kind='hist', align='mid')

_ = ax.set_xticks(ticks=[0, 1, 2])
```

```python
ax = df_iris[['petal_width']].plot(kind='bar')

# sample every 5 ticks
ticks = ax.get_xticks()[::5]
ax.set_xticks(ticks)
_ = ax.set_xticklabels(ticks) # important, otherwise the label values would be the index values of ticks
```
