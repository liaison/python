## Abstract

In this notebook, we show some examples on how to use the python libraries notably matplotlib and seaborn to plot data from DataFrame.

```python
import pandas as pd
import numpy as np

from sklearn import datasets

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = datasets.load_iris()

df_iris = pd.DataFrame(iris.data, columns=iris.feature_names)
df_iris['target'] = iris['target']

df_iris.head()
```

```python
import seaborn as sns

sns.pairplot(df_iris)
```

```python
_ = sns.pairplot(df_iris, hue='target')
```

```python
sns.set()

sns.pairplot(df_iris, hue='target')
_ = plt.suptitle('Scatter Plot for Iris data')
```

```python
sns.set(style="whitegrid", palette="muted")

sns.pairplot(df_iris, diag_kind='kde')
```
