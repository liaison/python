#### Decision Tree

Often one applies the decision tree classifier/regressor within the boosting ensemble. Reference: [the sklearn documentation](http://scikit-learn.org/stable/modules/tree.html)

```python
from sklearn import tree

X = [[0, 0], [1, 1]]
Y = [0, 1]

clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, Y)

print('Binary prediction result:', clf.predict([[2, 2]]))

print('Prediction with probability:', clf.predict_proba([[2, 2]]))
```

```python
from sklearn.datasets import load_iris

iris= load_iris()
clf = tree.DecisionTreeClassifier()
clf = clf.fit(iris.data, iris.target)
```

```python
import graphviz

# to visualize the decision tree
dot_data = tree.export_graphviz(clf, out_file=None, 
                         feature_names=iris.feature_names,  
                         class_names=iris.target_names,  
                         filled=True, rounded=True,  
                         special_characters=True)  

graph = graphviz.Source(dot_data)

graph 
```

```python
print(iris.data[:1, :], clf.predict(iris.data[:1, :]))
print(iris.data[:1, :], clf.predict_proba(iris.data[:1, :]))
```
